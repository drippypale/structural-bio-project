# Reading $\phi, \psi$ angles from the `phi_psi.txt` file:

```python
with open('phi_psi.txt') as f:
    phi, psi = [], []
    for line in f:
        phi.append(float(line.split()[0]))
        psi.append(float(line.split()[1]))
```

## Removing first and last angle:

Since phi for the first residue and psi for the last residue are not defined.

```python
phi = phi[1:-1]
psi = psi[1:-1]
```

## Create a `DataFrame` with the angles:

```python
import pandas as pd
```

```python
df = pd.DataFrame({'phi': phi, 'psi': psi})
```

# Plotting the *Ramachandran plot*:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(font_scale=1.5)

ax = sns.scatterplot(data=df, x='phi', y='psi', alpha=1).set(xlabel='$\phi$', ylabel='$\psi$', xticklabels=[], yticklabels=[])

plt.title('Ramachandran Plot', y=1.1)

plt.show()
```
